```python
import pandas as pd
col_names = ["pregnant","glucose","bp","skin","insulin","bmi","pedigree","age","outcome"]
df = pd.read_csv("diabetes.csv",names = col_names).iloc[1:]
print(df.head())
```

```python
features = ["pregnant","insulin","bmi","age","glucose","bp","pedigree"]

X = df[features]
Y = df.outcome
```

```python
from sklearn.model_selection import train_test_split as tts
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

x_train,x_test,y_train,y_test = tts(X,Y,test_size=0.3,random_state = 1)

clf = DecisionTreeClassifier()

clf = clf.fit(x_train,y_train)

y_pred = clf.predict(x_test)

print("Accuracy score is ",accuracy_score(y_test,y_pred))

```

```python
from sklearn.tree import export_graphviz
from io import StringIO
import pydotplus

dot_data = StringIO()
export_graphviz(clf,out_file = dot_data,filled = True,special_characters = True,rounded = True,feature_names = features,class_names = ['0','1'])

print(dot_data.getvalue())
```

```python
from IPython.display import Image

graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png("diabetes.png")
Image(graph.create_png())
```

```python
clf = DecisionTreeClassifier(max_depth = 3)
clf = clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)

print("Model accuracy is : ", accuracy_score(y_test, y_pred))
```

```python
dot_data = StringIO()
export_graphviz(clf, out_file = dot_data, filled = True, rounded = True, special_characters = True, feature_names = features, class_names = ['0', '1'])

graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png("Diabetes.png")
Image(graph.create_png())
```
